```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification

X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
model.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
score=accuracy_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
report=classification_report(y_test,y_pred)

print(score)
print(cm)
print(report)
```

**Hyper Parameter Tuning and Cross Validation**

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
params=dict(penalty=penalty,C=c_values)
params
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

**GRID SEARCH CV**

```python
# GridsearchCV
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=params,cv=cv,n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

**RANDOMIZED SEARCH CV**

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

**IMBALANCED DATASET**

```python
# Generate and plot a synthetic imbalanced classification dataset
from collections import Counter
from sklearn.datasets import make_classification
```

```python
# Imbalanced dataset
X,y=make_classification(n_samples=10000,n_features=2,n_clusters_per_class=1,
                        n_redundant=0,weights=[0.99],random_state=10)
```

```python
Counter(y)
```

```python
X
```

```python
y
```

```python
# Convert X to a DataFrame for easier plotting
import seaborn as sns
X = pd.DataFrame(X, columns=["Feature_1", "Feature_2"])

# Plot the data
sns.scatterplot(x=X["Feature_1"], y=X["Feature_2"], hue=y,palette='viridis')
```

```python
# Splitting the dataset
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
class_weight=[{0:w,1:y}for w in [1,10,50,100]for y in [1,10,50,100]]
```

```python
class_weight
```

```python
# Hyperparameter tuning
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','lbfgs','liblinear','sag','saga']
class_weight=[{0:w,1:y}for w in [1,10,50,100]for y in [1,10,50,100]]
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver,class_weight=class_weight)
params
```

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
cv=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,cv=cv,scoring='accuracy')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
y_pred
```

```python
score=accuracy_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
report=classification_report(y_test,y_pred)

print(score)
print(cm)
print(report)
```

**Logistic Regression with ROC curve and AUC score**

```python
from sklearn.metrics import roc_curve,roc_auc_score
from matplotlib import pyplot
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#Create a dummy model with default 0 as output
dummy_model_prob=[0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model_prob=model.predict_proba(X_test)
model_prob
```

```python
model_prob=model_prob[:,1]
model_prob
```

```python
dummy_model_auc=roc_auc_score(y_test,dummy_model_prob)
print(dummy_model_auc)
model_auc=roc_auc_score(y_test,model_prob)
print(model_auc)
```

```python
# calculate roc curves
fpr,tpr,_=roc_curve(y_test,dummy_model_prob)
fpr1,tpr1,_=roc_curve(y_test,model_prob)
print(fpr,tpr)
print(fpr1,tpr1)
```

```python
# plot the roc curve for model
pyplot.plot(fpr,tpr,linestyle='--',label='Dummy Model')
pyplot.plot(fpr1,tpr1,marker='.',label='Logistic Regression')
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
pyplot.legend()
pyplot.show()
```
